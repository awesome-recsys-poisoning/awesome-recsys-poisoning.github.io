```python
import re

def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
f = open("output.txt", "w")
```

```python
for line in lines:
    if line=="Model-Agnostic\n" or line=="Model-Intrinsic\n" or line=="Data-Driven\n":
        types=line[:-1]
    else:
        print('----')
        ### title
        title = re.search(r'\[(.*?)\]', line).group()[1:-1]

        ### link to paper
        link = re.search(r'\((.*?)\)', line).group()[1:-1]

        try:
        ### venue
            venue = re.search(r'\|\s\_(.*?)\_', line).group()[3:-1]
            print("re.search(r'\|\s\_(.*?)\_', line)", re.search(r'\|\s\_(.*?)\_', line))
        except:
            match = re.search(r'\|\s*\[.*?\]\(.*?\)\s*\|\s*(\w+)\s*\|', line)

            if match:
                print("Found:", match.group(1))
                venue = match.group(1)
            else:
                print("No match found.")

        ### year
        year = re.search(r'\s\d{4}\s', line).group()[1:-1]

        ### code
        res = re.search(r'\|\s*\[.*?\]\(.*?\)\s*\|\s*\w+\s*\|\s*\d{4}\s*\|\s*(.*?)\s*\|', line)
        if res is None:
            code = '-'
        else:
            code = res.group(1)

#         ### type
#         type_slash_idx = findOccurrences(line, '|')[-2:]
#         types = line[type_slash_idx[0]:type_slash_idx[1]][2:-1].lstrip()

        print(title)
        print(link)
        print(venue)
        print(year)
        print(code)
        print(types)

        ### formation
        f.write("                                                <tr>")
        f.write("\n")
        f.write("                                                    <td><a href=\"{}\">{}</a></td>".format(link, title))
        f.write("\n")
        f.write("                                                    <td>{}</td>".format(venue))
        f.write("\n")
        f.write("                                                    <td>{}</td>".format(year))
        f.write("\n")
        f.write("                                                    <td>{}</td>".format(code))
        f.write("\n")
        f.write("                                                </tr>")
    f.write("\n")
    
f.close()
```
